```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge,Lasso
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing=fetch_california_housing(as_frame=True)
```

```python
df=housing.frame
```

```python
df.head()
```

```python
x=df.iloc[:,:8]
```

```python
x
```

```python
y=df.iloc[:,8:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr=LinearRegression()
```

```python
mlr.fit(x_train,y_train)
```

```python
y_pred=mlr.predict(x_test)
```

```python
y_test.values
```

```python
y_pred
```

```python
mlr.coef_
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Model Performance:")
print("Mean Squared Error (MSE):", mse)
print("RÂ² Score:", r2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred,alpha=0.5,color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler= StandardScaler()
```

```python
X_train_scaled = Scaler.fit_transform(x_train)
X_test_scaled = Scaler.transform(x_test)
```

```python
X_train_scaled
```

```python
x_train.values
```

```python
mlr = LinearRegression()
mlr.fit(X_train_scaled, y_train)
y_pred = mlr.predict(X_test_scaled)
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_test_scaled)
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_test_scaled)
```

```python
mse_mlr=mean_squared_error(y_pred,y_test)
mse_ridge =mean_squared_error(y_pred_ridge,y_test)
mse_lasso =mean_squared_error(y_pred_lasso,y_test)
print("MSE of MLR:",mse_mlr)
print("MSE of RIDGE:",mse_ridge)
print("MSE of LASSO:",mse_lasso)
```

```python
R2_mlr=r2_score(y_test, y_pred)
R2_ridge =r2_score(y_test, y_pred_ridge)
R2_lasso =r2_score(y_test, y_pred_lasso)
print("r2_score of MLR:",R2_mlr)
print("r2_score  of RIDGE:",R2_ridge)
print("r2_score  of LASSO:",R2_lasso)
```

```python
mlr = ['Linear', 'Ridge', 'Lasso']
mse_values = [
    mean_squared_error(y_test, y_pred),
    mean_squared_error(y_test, y_pred_ridge),
    mean_squared_error(y_test, y_pred_lasso)
]
r2_values = [
    r2_score(y_test, y_pred),
    r2_score(y_test, y_pred_ridge),
    r2_score(y_test, y_pred_lasso)
]

result_df = pd.DataFrame({
    'Model': mlr,
    'MSE': mse_values,
    'RÂ² Score': r2_values
})
print(result_df)
```
